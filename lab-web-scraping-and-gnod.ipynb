```python
sp_songs_df = pd.read_csv('spotify_songs_b.csv')
```

```python
print(sp_songs_df.info())
sp_songs_df.head()
```

```python
#select only numerical features from the dataframe to be used in K-means model

songs_num_df = sp_songs_df.select_dtypes(np.number)
songs_num_df
```

```python
# scaling the numerical df using StandardScaler()

Standardtransformer = StandardScaler().fit(songs_num_df)     # this transformer will be used later
X_prep = Standardtransformer.transform(songs_num_df)         # scaling
X_prep = pd.DataFrame(X_prep,columns=songs_num_df.columns)   # creating pd.df
X_prep
```

```python
type(X_prep)
```

```python
# pickle saving scaler

pickle.dump(Standardtransformer, open('Standardtransformer.pkl', 'wb'))
```

```python
# building K-means clustering models to choose optimal number of K

K = range(2, 70)
inertia = []

for k in K:
    
    kmeans = KMeans(n_clusters=k, n_init='auto', random_state=200)
    kmeans.fit(X_prep)
    inertia.append(kmeans.inertia_)

%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

The inertia chart suggests (I guess) the optimal value of K may be 15.

```python
# the model

kmeans = KMeans(n_clusters=42, n_init='auto', random_state=200)
kmeans.fit(X_prep)
```

```python
# pickle saving kmeans model

pickle.dump(kmeans, open('kmean.pkl', 'wb'))
```

```python
# Predicting / assigning the clusters:

clusters = kmeans.predict(X_prep)
print(len(clusters))
pd.Series(clusters).value_counts().sort_index()
```

```python
# a new column for the category, into the sp_songs_df

sp_songs_df['cluster'] = clusters
sp_songs_df.head()
```

```python
# saving the sp_songs_df updated w/cluster column to a .csv file

sp_songs_df.to_csv('sp_songs_df.csv', index=False)
```
